# EJERCICIOS USANDO DATASET TITANIC

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset Titanic
titanic = sns.load_dataset("titanic")
titanic.dropna(inplace=True)
titanic
```

# Visualización con Seaborn en dataset_Titanic

---

## Visualización de Distribución con `displot`





```python
sns.displot(data=titanic, x="age", bins=30, kde=True)
plt.title("Distribución de edades de los pasajeros del Titanic")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

## Relación entre dos variables numéricas con `scatterplot`





```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="sex", alpha=0.6)
plt.title("Relación entre Edad y Tarifa según Sexo")
plt.xlabel("Edad")
plt.ylabel("Tarifa")
plt.show()
```

## Comparación categórica con `catplot`



```python
sns.catplot(
    data=titanic,
    x="class",
    hue="sex",
    kind="count",
    col="survived"
)
```

---

## Análisis bivariado con `jointplot`

*

```python
sns.jointplot(data=titanic, x="age", y="fare", kind="kde", height=5)
plt.show()
```

## Matriz de relacion con `pairplot`



```python
sns.pairplot(data=titanic, vars=["age", "fare", "pclass"], hue="sex", corner=True)
plt.show()
```
